<h1 style ='font-size:40px; text-align: center; font-weight:500; color:#f5d6e3'>Income Prediction Based on Demographic and Socio-Economic Factors</h1>

### ***Objective:***
To analyze and predict an individual's income per year based on various demographic and socio-economic factors.

### ***Goal:***
Using this analysis, the aim is to build a predictive model that can accurately income of an individual person per year based on the given features.

This analysis will help in identifying key factors that influence higher income levels and provide insights for policies or strategies to enhance economic well-being.

**Dataset:** [Link](https://github.com/Satwik-uppada)

### ***Dataset Description:***

The dataset contains demographic and socio-economic information for individuals. The goal is to predict whether an individual's income exceeds $50K per year. The dataset includes the following features:

- **age:** Age of the individual
- **workclass:** Employment type (e.g., Private, Self-emp-not-inc)
- **fnlwgt:** Final weight, indicating the number of people the census believes the entry represents
- **education:** Highest level of education achieved
- **educational-num:** Number corresponding to the level of education
- **marital-status:** Marital status (e.g., Married, Never-married)
- **occupation:** Type of occupation
- **relationship:** Relationship status (e.g., Husband, Not-in-family)
- **race:** Race of the individual (e.g., Black, white)
- **gender:** Gender of the individual (e.g., Male, Female)
- **capital-gain:** Capital gains from investment
- **capital-loss:** Capital losses from investment
- **hours-per-week:** Average hours worked per week
- **native-country:** Country of origin
- **income:** Income level (e.g.,>$50K or <=$50K)
- **Income:** Salary of that individual
