```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
import pandas as pd
from bs4 import BeautifulSoup
import time
import os

cwd = os.getcwd()

# Specify the path to the Edge WebDriver executable file
driver_path = f"{cwd}\\edgedriver_win64\\msedgedriver.exe"

# Create a Service object with the executable path
service = Service(driver_path)

# Set up the Edge WebDriver using the Service object
driver = webdriver.Edge(service=service)

# Define the URL
link = "https://www.cars24.com/buy-used-car?f=make%3A%3D%3Amaruti%3AOR%3Amake%3A%3D%3Ahonda%3AOR%3Amake%3A%3D%3Atata&sort=bestmatch&serveWarrantyCount=true&gaId=1279799800.1722605877&storeCityId=2"


# Function to scroll and load page content
def scroll_and_load_page(driver, scroll_pause_time=2):
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause_time)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# Navigate to the Amazon.in best sellers page
driver.get(link)

# Wait for the page to load and perform scrolling
time.sleep(5)
scroll_and_load_page(driver)

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# First data extraction: Car names and years
car_name_year_elements = soup.find_all('h3', {'class': '_11dVb'})
car_name_year = [car.text for car in car_name_year_elements]

name = []  # spliting the car names from car_name_year and appending in name list
year = [] # spliting the year of registration from car_name_year and appending in year list

for i in car_name_year:
    car_year, car_name = i.split(" ",1)
    name.append(car_name)
    year.append(car_year)
    
# Reload the page to perform the second extraction
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)



# -----> second data extraction: Km driven 
# Parse the page source again with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Second data extraction: Kilometers
km_elements = soup.find_all('ul', {'class': '_3J2G-'})
km = []

for i in km_elements:
    lis = i.find_all("li")
    if len(lis) >= 2:
        sec_li = lis[0]
        text = sec_li.get_text(strip=True)
        km.append(text if text else 'N/A')

# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)




# -----> Third data extraction: fuel type

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('ul', {'class': '_3J2G-'})
fuel_type = []

for i in data:
    lis = i.find_all("li")
    
    if len(lis) >=2:
        sec_li = lis[2]
        text = sec_li.get_text(strip=True)
        fuel_type.append(text if text else 'N/A')


# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)



# -----> fourth data extration: transmission type

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('ul', {'class': '_3J2G-'})
transmission_type = []

for i in data:
    lis = i.find_all("li")
    
    if len(lis) >=2:
        sec_li = lis[4]
        text = sec_li.get_text(strip=True)
        transmission_type.append(text if text else 'N/A')


# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)



# -----> fifth data extraction: Discount price

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('strong', {'class': '_3RL-I'})
discount_price = []

for i in data:
    discount_price.append(i.get_text() if i else 'N/A')
    

# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)




# -----> sixth data extraction: actual price
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('div', {'class': '_2KyOK'})
actual_price = []
# 
for i in data:
    span = i.find("span")
    # 
    actual_price.append(span.get_text() if span else 'N/A')
    

# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)



# -----> Seventh data extraction: discount offered
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('div', {'class': '_2KyOK'})
discount = []
# 
for i in data:
    span = i.find('span',{'class': 'YWP1x'})
    discount.append(span.get_text() if span else 'N/A')
    

# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)


# -----> Eighth data extration: model
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('div', {'class': '_2YB7p'})
add_ons = []

for i in data:
    span = i.find("span",{'class':'_3JoYA'})
    add_ons.append(span.get_text() if span else 'N/A')


# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)




# ----->  last attribute : test drive location 
test_drive_loc = []
for i in soup:
    drive_location_div = i.find_all("div",{'class':'_2YB7p'})
    for i in drive_location_div:
        if i.find("p",{'class':'_3dGMY'}):
            p_tag = i.find_all("p",{'class':'_3dGMY'})
            for tag in p_tag:
                spans = tag.find_all("span")
                span = spans[1]
                if span:
                    test_drive_loc.append(span.get_text())
    else:
        tokens = soup.find_all("span",{'class':'_1MKmC'})
        for token in tokens:
            test_drive_loc.append(token.get_text())

    
# close the driver
driver.quit()


# feature spliting for brand and model from name list 
brand = []
car_model = []
for i in name:
    brand_name, model_name = i.split(" ",1)
    brand.append(brand_name)
    car_model.append(model_name)

# last data extraction : location
location = [soup.find("div",{'class':'_2SBi0'}).get_text()] * len(name)
```

```python
link = "https://www.cars24.com/buy-used-car?f=make%3A%3D%3Atata&sort=bestmatch&serveWarrantyCount=true&gaId=1279799800.1722605877&listingSource=TabFilter&storeCityId=2378"

from selenium import webdriver
from selenium.webdriver.edge.service import Service
import pandas as pd
from bs4 import BeautifulSoup
import time
import os

cwd = os.getcwd()

# Specify the path to the Edge WebDriver executable file
driver_path = f"{cwd}\\edgedriver_win64\\msedgedriver.exe"

# Create a Service object with the executable path
service = Service(driver_path)

# Set up the Edge WebDriver using the Service object
driver = webdriver.Edge(service=service)

# Define the URL
# link = "https://www.cars24.com/buy-used-car?f=make%3A%3D%3Amaruti%3AOR%3Amake%3A%3D%3Ahonda%3AOR%3Amake%3A%3D%3Atata&sort=bestmatch&serveWarrantyCount=true&gaId=1279799800.1722605877&storeCityId=2"


# Function to scroll and load page content
def scroll_and_load_page(driver, scroll_pause_time=2):
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause_time)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# Navigate to the Amazon.in best sellers page
driver.get(link)

# Wait for the page to load and perform scrolling
time.sleep(5)
scroll_and_load_page(driver)

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# First data extraction: Car names and years
car_name_year_elements = soup.find_all('h3', {'class': '_11dVb'})
car_name_year = [car.text for car in car_name_year_elements]

name = []  # spliting the car names from car_name_year and appending in name list
year = [] # spliting the year of registration from car_name_year and appending in year list

for i in car_name_year:
    car_year, car_name = i.split(" ",1)
    name.append(car_name)
    year.append(car_year)
    
# Reload the page to perform the second extraction
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)



# -----> second data extraction: Km driven 
# Parse the page source again with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Second data extraction: Kilometers
km_elements = soup.find_all('ul', {'class': '_3J2G-'})
km = []

for i in km_elements:
    lis = i.find_all("li")
    if len(lis) >= 2:
        sec_li = lis[0]
        text = sec_li.get_text(strip=True)
        km.append(text if text else 'N/A')

# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)




# -----> Third data extraction: fuel type

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('ul', {'class': '_3J2G-'})
fuel_type = []

for i in data:
    lis = i.find_all("li")
    
    if len(lis) >=2:
        sec_li = lis[2]
        text = sec_li.get_text(strip=True)
        fuel_type.append(text if text else 'N/A')


# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)



# -----> fourth data extration: transmission type

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('ul', {'class': '_3J2G-'})
transmission_type = []

for i in data:
    lis = i.find_all("li")
    
    if len(lis) >=2:
        sec_li = lis[4]
        text = sec_li.get_text(strip=True)
        transmission_type.append(text if text else 'N/A')


# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)



# -----> fifth data extraction: Discount price

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('strong', {'class': '_3RL-I'})
discount_price = []

for i in data:
    discount_price.append(i.get_text() if i else 'N/A')
    

# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)




# -----> sixth data extraction: actual price
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('div', {'class': '_2KyOK'})
actual_price = []
# 
for i in data:
    span = i.find("span")
    # 
    actual_price.append(span.get_text() if span else 'N/A')
    

# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)



# -----> Seventh data extraction: discount offered
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('div', {'class': '_2KyOK'})
discount = []
# 
for i in data:
    span = i.find('span',{'class': 'YWP1x'})
    discount.append(span.get_text() if span else 'N/A')
    

# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)


# -----> Eighth data extration: model
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all elements with the specified class
data = soup.find_all('div', {'class': '_2YB7p'})
add_ons = []

for i in data:
    span = i.find("span",{'class':'_3JoYA'})
    add_ons.append(span.get_text() if span else 'N/A')


# refresh the driver
driver.refresh()

# Wait for the page to reload and perform scrolling again
time.sleep(5)
scroll_and_load_page(driver)




# ----->  last attribute : test drive location 
test_drive_loc = []
for i in soup:
    drive_location_div = i.find_all("div",{'class':'_2YB7p'})
    for i in drive_location_div:
        if i.find("p",{'class':'_3dGMY'}):
            p_tag = i.find_all("p",{'class':'_3dGMY'})
            for tag in p_tag:
                spans = tag.find_all("span")
                span = spans[1]
                if span:
                    test_drive_loc.append(span.get_text())
    else:
        tokens = soup.find_all("span",{'class':'_1MKmC'})
        for token in tokens:
            test_drive_loc.append(token.get_text())

    
# close the driver
driver.quit()


# feature spliting for brand and model from name list 
brand = []
car_model = []
for i in name:
    brand_name, model_name = i.split(" ",1)
    brand.append(brand_name)
    car_model.append(model_name)

# last data extraction : location
location = [soup.find("div",{'class':'_2SBi0'}).get_text()] * len(name)

assert len(brand) == len(car_model) == len(year) == len(km) == len(fuel_type) == len(transmission_type) == len(add_ons) == len(discount_price) == len(actual_price) == len(discount) == len(location) == len(test_drive_loc)
data = list(zip(brand, car_model, year, km, fuel_type, transmission_type, add_ons, discount_price, actual_price, discount, location, test_drive_loc))
columns = ['Brand', 'Model', 'Registered Year', 'Km Driven', 'Fuel Type', 'Transmission Type', 'Add ONs', 'Offer Price','Original Price','Discount Offered','Location','Test Drive Location']
homya_data_1 = pd.DataFrame(data, columns=columns)
homya_data_1
```

```python
def convert_to_float(price_str):
    if isinstance(price_str, str):
        # Remove rupee symbol and "Lakh" text, and convert to float
        price_str = price_str.replace('₹', '').replace(' Lakh', '').strip()
        return float(price_str) * 100000
    return None

# Apply the conversion function
merged_data['Original Price'] = merged_data['Original Price'].apply(convert_to_float)
```

```python
def convert_discount(discount_str):
    if isinstance(discount_str, str):
        # Remove parentheses and text
        discount_str = discount_str.replace('(', '').replace(')', '').replace(' off', '').strip()
        
        # Handle thousands and lakhs
        if 'L' in discount_str:
            discount_str = discount_str.replace('L', '').strip()
            return float(discount_str) * 100000  # Convert lakhs to actual number
        elif 'k' in discount_str:
            discount_str = discount_str.replace('k', '').strip()
            return float(discount_str) * 1000  # Convert thousands to actual number
        else:
            return float(discount_str)  # Already in actual number
    return None

# Apply the conversion function
merged_data['Discount Offered'] = merged_data['Discount Offered'].apply(convert_discount)
```

```python
# Function to clean and convert km driven
def convert_km(km_str):
    if isinstance(km_str, str):
        # Remove ' km' and commas
        km_str = km_str.replace(' km', '').replace(',', '')
        return int(km_str)  # Convert to integer
    return pd.NA

# Apply the conversion function
merged_data['Km Driven'] = merged_data['Km Driven'].apply(convert_km)
```
