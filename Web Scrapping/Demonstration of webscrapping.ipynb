```python
import requests
result = requests.get("http://www.imdb.com/search/title?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012")
```

```python
import bs4
soup = bs4.BeautifulSoup(result.text,'html')
print(soup)
```

```python
import bs4
soup = bs4.BeautifulSoup(result.text,'html.parser')
print(soup)
```

### body of the Webpage

```python
cases = soup.find_all("div",class_="lister-item mode-advanced")
print(len(cases))
print(cases)
```

### Extracting the data from webpage

```python
year = []
Run = []
name= []
rating = []
genre=[]
certificate =[]
gross_votes_list = []
meta_score = []

for i in cases:
    year_mov = i.find("span",class_="lister-item-year")
    year.append(year_mov.string)
    Runtime_mov = i.find("span",class_="runtime")
    Run.append(Runtime_mov.string)
    name_mov = i.find('img')['alt']
    name.append(name_mov)
    rating_mov= i.find("strong")
    rating.append(rating_mov.text)
    genre_mov = i.find("span",class_="genre")
    genre.append(genre_mov.get_text(strip=True))
    mov_certificate = i.find("span",class_="certificate")
    certificate.append(mov_certificate.text)
    meta_score_mov=i.find("div",class_="inline-block ratings-metascore")
    meta_score.append(meta_score_mov.get_text(strip=True).replace("Metascore",""))
    spans = i.find_all("span",{"name":"nv"})
    for span in spans:
        gross_votes_list.append(span.text)
    
gross_list=gross_votes_list[1::2]
votes_list=gross_votes_list[::2]
print(gross_list)
print(votes_list)
print("Year: ",year)
print("Runtime: ",Run)
print("Movie Name: ",name)
print("Rating: ",rating)
print("Genre: ",genre)
print("Certificate: ",certificate)
print("Gross: ",gross_list)
print("Votes: ",votes_list)
print("MetaScore: ",meta_score)

 
```

### Creating data frame for the extracted data

```python
import pandas as pd
dataframe = pd.DataFrame({'Name':name,'Year':year,'Runtime':Run,"Rating":rating,"Certificate":certificate,"Genre":genre,"Gross":gross_list,"Votes":votes_list,"MetaScore":meta_score})
pd.set_option('display.expand_frame_repr', False)
print(dataframe)
```
