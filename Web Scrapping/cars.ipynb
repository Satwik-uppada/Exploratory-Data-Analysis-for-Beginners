
# Acko Drive Webscrapping 

## Importing libraries

```python
import pandas as pd 
from bs4 import BeautifulSoup
import requests
```

## storing website link in a variable

```python
link = "https://ackodrive.com/collection/tata-cars/"
```

## Requesting data from HTTPS

```python
response = requests.get(link)
response.status_code
```

## Soup Object

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

## Name

```python
results = soup.find_all('h2',{'class':'styles__ModelTitle-sc-57d31ed8-4 LQrbf'})
name_list = []
for i in results:
    name_list.append(i.get_text())
```

## Millage

```python
fuel_type = soup.find_all('p',{'data-testid':'car_variant_fuel_type'})
fuel_list = []
for i in fuel_type:
    fuel_list.append(i.get_text())
```

```python

transmission_type = soup.find_all('p',{'data-testid':"car_variant_transmission"})
transmission_list = []
for i in transmission_type:
    transmission_list.append(i.get_text())
```

```python


colors = soup.find_all('p',{'class':"styles__ParaWithoutMargins-sc-57d31ed8-33 jvhnwy"})
colors_list = []

for i in range(5,len(colors),6):
    colors_list.append(colors[i].get_text())
```

```python

price = soup.find_all('div',{'data-testid':'car_price'})
price_list = []

for i in price:
    price_list.append(i.get_text())
```

```python
varient = soup.find_all('h4',{'class':"styles__VariantTitle-sc-57d31ed8-10 haSBAh"})
varient_list = []
for i in varient:
    varient_list.append(i.get_text())
```

```python


model = soup.find_all('p',{'data-testid':"car_model_body_type"})
model_list = []
for i in model:
    model_list.append(i.get_text())
```

```python
varients_count = soup.find_all('p',{'data-testid':"car_model_variants"})
varient_count_list = []
for i in varients_count:
    varient_count_list.append(i.get_text())
```

```python


seats = soup.find_all('p',{'data-testid':"car_model_seat"})
seats_list = []
for i in seats:
    seats_list.append(i.get_text())

```

```python

city = soup.find_all('div',{'class':"styles__CityName-sc-57d31ed8-17 lhXIXl"})
city_list = []
for i in city:
    city_list.append(i.get_text())

```

```python
assert len(name_list) == len(fuel_list) == len(transmission_list) == len(colors_list) == len(price_list) == len(varient_list) == len(varient_count_list) == len(seats_list) ==len(city_list)
```

```python
import pandas as pd

data = list(zip(name_list, fuel_list, transmission_list, colors_list, price_list, varient_list, varient_count_list, seats_list, city_list))
columns = ['Car Name', 'Fuel Type', 'Transmission Type', 'Available Colors', 'Price', 'Variant', 'Number of Variants Available', 'Seater','city']


df = pd.DataFrame(data, columns=columns)

df
```
