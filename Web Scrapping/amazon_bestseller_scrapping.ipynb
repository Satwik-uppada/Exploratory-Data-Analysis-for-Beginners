```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url ="https://www.amazon.in/gp/bestsellers/kitchen/1379960031/ref=zg_bs_nav_kitchen_1"

```

```python
response = req.get(url)
print(response)
```

```python
soup = bs(response.text,parser="lxml")
soup
```

```python
content = soup.find_all("div",{"class":"p13n-gridRow _cDEzb_grid-row_3Cywl"})

hash_tags_list = []
for i in content:

    hash_tags = i.find_all("span",{"class":"zg-bdg-text"})
    for i in hash_tags:
     hash_tag = i.get_text(strip=True) if i else "N/A"
     hash_tags_list.append(hash_tag)

    name = i.find_all("div",{"class":"zg-grid-general-faceout"})
    print(name)
   
   
```

```python
content_2 = soup.find_all("div",{'class':'_cDEzb_p13n-sc-css-line-clamp-3_g3dy1'})
names_list = []
for i in content_2:
    hash_tag = i.get_text(strip=True) if i else "N/A"
    names_list.append(hash_tag)

names_list
```
