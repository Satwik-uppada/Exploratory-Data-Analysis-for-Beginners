<h3>Importing requested modules for webscrapping</h3>

```python
import requests as req
from bs4 import BeautifulSoup as bs
import numpy as np
import statistics as stat
import pandas as pd
```

## URLS 

```python
Base_URL ="https://books.toscrape.com/"
url = Base_URL + "catalogue/category/books/science-fiction_16/index.html"
```

### Requesting the url 

```python
response = req.get(url)
print(response)
```

### Fetching html code of the url

```python
soup = bs(response.text,parser = "lxml")
print(soup)
```

### Trying to print all the paragraphs

```python
print(soup.find_all("p"))
```

### Trying get all the anchor tags

```python
print(soup.find_all("a"))
```

### Printing the prices of the books

```python
price_tags =  soup.find_all("p",{"class":"price_color"})
prices =[float(price.text[2:]) for price in price_tags]
print(prices)
```

### mean of that prices

```python
print(stat.mean(prices))
```

### Status of Books Availablity

```python
books_availability = soup.find_all("p",{"class":"instock availability"})
available_status = ["In stock" if "In stock" in text.get_text(strip=True) else "Currently Unavailable" for text in books_availability]
print(available_status)
```

### Names of Books

```python
names = soup.find("ol").find_all('h3')
content =[]
for x in names:
    content.append(x.get_text())

print(content)
```

### Rating of the books

```python
rating_element = soup.find('ol').find_all("p",class_ ="star-rating")
rating = []
for star_rating in rating_element:
 if star_rating['class'][1] == "One":
    rating.append("*")
 elif star_rating['class'][1] == "Two":
    rating.append("* *")
 elif star_rating['class'][1] == "Three":
    rating.append("* * *")
 elif star_rating['class'][1] == "Four":
    rating.append("* * * *")
 elif star_rating['class'][1] == "Five":
    rating.append("* * * * *")
 else:
    rating.append()

print(rating)
```

### Creating data frame for the above content

```python
data = pd.DataFrame({"Book Title":content,"Price":prices,"Availablity_Status":available_status,"Rating of the book":rating})
pd.set_option('display.expand_frame_repr', False)

print(data)
```
